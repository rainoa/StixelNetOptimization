```python
import numpy as np
import pandas as pd
import math
import timeit
import matplotlib.pyplot as plt
from matplotlib.image import imread
import os
from os.path import expanduser
%matplotlib inline
```

```python
percent = 2
img_path = os.path.join('..','datasets','stixels')
sum_csv_path = os.path.join(img_path,'train', 'sum_csv')
labels_train_df = pd.read_csv(os.path.join(sum_csv_path,'labels_train_'+str(percent)+'percent.csv'))
```

```python
labels_train_df.head()
```

```python
labels_tr = labels_train_df['Label'].values.tolist()
print(labels_tr)
```

```python
len(labels_tr)/64
```

```python
bins_edges = [x for x in range(0, 47, 1)]
hist_res_tr = np.histogram(labels_tr, bins=bins_edges)
```

```python
plt.rcParams['figure.figsize'] = [10, 5]
plt.bar(bins_edges[:-1], height=hist_res_tr[0])
plt.xlabel('Label')
plt.ylabel('# examples')
plt.title('Distribution of obsticle location in stixel - train')
plt.rcParams.update({'axes.labelsize': 'x-large'})
plt.rcParams.update({'axes.titlesize': 'x-large'})
```

```python
#val
sum_csv_path = os.path.join(img_path,'val', 'sum_csv')
labels_val_df = pd.read_csv(os.path.join(sum_csv_path,'labels_val_'+str(percent)+'percent.csv'))
labels_val = labels_val_df['Label'].values.tolist()
hist_res_val = np.histogram(labels_val, bins=bins_edges)
```

```python
plt.bar(bins_edges[:-1], height=hist_res_val[0])
plt.xlabel('Label')
plt.ylabel('# examples')
plt.title('Distribution of obsticle location in stixel - validation')
```
